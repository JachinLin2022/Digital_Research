```python

def calculate_metric(row):
    pos = row['pos']
    neg = row['neg']
    neu = row['neu']
    
    if pos >= neg and pos >= neu:
        return 1
    elif neg >= pos and neg >= neu:
        return -1
    else:
        return 0

sentiment['metrics_tmp'] = sentiment.apply(calculate_metric, axis=1)
sentiment
```
