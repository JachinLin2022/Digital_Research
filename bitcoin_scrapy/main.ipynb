```python
import os
import re
import pandas as pd


source_id = pd.read_csv('update/0729/all_filter_update_onlyid.csv',dtype={'id':str})
print(source_id)
files = os.listdir('tweet/new_day')

res = ''
for year in ['2023']:

    all = 0
    end = 0
    for idx,file in enumerate(files):
        if file.find(year) < 0:
            continue
        tweets = pd.read_csv('tweet/new_day/' + file, dtype={'id':str})
        print(file)

        if len(tweets) == 0:
            continue
        if len(res) == 0:
            res = tweets
        else:
            res = pd.concat([res,tweets])


res.drop_duplicates(subset=['id'], keep='first', inplace=True)
res = res[~res['id'].isin(source_id['id'])]
print(res)
```

```python
res['second'] = pd.to_datetime(res['created_at'], format='%a %b %d %H:%M:%S +0000 %Y')
res['month'] = res['second'].dt.year.astype(str) + '-' + res['second'].dt.month.astype(str)
res['day'] = res['second'].dt.year.astype(str) + '-' + res['second'].dt.month.astype(str) + '-' + res['second'].dt.day.astype(str)
res['second'] = res['second'].astype('int64') // 10**9
res = res.sort_values('second', ascending=True)

import re
def remove_links(text):
    # 匹配链接的正则表达式
    pattern = r'http\S+|www\S+'
    # 使用空字符串替换链接
    text_without_links = re.sub(pattern, '', text)
    return text_without_links
# def is_english(text):
#     import string
#     """判断文本是否为英文"""
#     # 忽略掉单词数等于0的文本
#     if len(text) == 0:
#         return False
#     english_letters = set(string.ascii_letters)

#     # 统计文本中英文字母的比例
#     english_letters = [char for char in text if char.isalpha() and char in english_letters]
#     english_ratio = len(english_letters) / len(text)
    
#     return english_ratio > 0.5
res['full_text'] = res['full_text'].astype(str)
res['full_text'] = res['full_text'].map(lambda x:remove_links(x))

res.drop_duplicates(subset=['id'], keep='first', inplace=True)
res.drop_duplicates(subset=['full_text'], keep='first', inplace=True)
print(res)
```
